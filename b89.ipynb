En primer lugar, eliminemos las columnas que no usaremos, ya que no afectarán el análisis.  


```python
#Eliminando Columnas que no utilizare en el análisis
columns_drop=['Unnamed: 0']
data.drop(columns_drop, axis =1 , inplace = True)

```

Ahora evalúo si no hay Valores nulos y duplicados.


```python
data[data.duplicated()]
```

```python
#Solo se tiene 1 valor duplicado, elimino uno y me quedo con el otro
data.drop_duplicates(keep='first',inplace=True)
```

```python
#Evaluo si hay Valores nulos
data.isnull().sum()
```

```python
#Muestro los registros con valores perdidos en el atributo
data[data.actual_price.isnull()]
```

```python
#Elimino los registros con valores nulos
data.dropna(subset = ["actual_price"], inplace=True)
data.dropna(subset = ["ratings"], inplace=True)
data.dropna(subset = ["discount_price"], inplace=True)
```

```python
data.isnull().sum()
```

```python
columnas = ['discount_price','actual_price']
data[columnas] = data[columnas].replace({',':''}, regex=True)
data[columnas] = data[columnas].replace({'₹':''}, regex=True)
```

```python
from deep_translator import GoogleTranslator
import pandas as pd

# Crear el objeto que nos permitirá hacer la traducción
translator = GoogleTranslator(source="en", target="es")

# Dividir la columna 'main_category' en una lista de textos
text_list = data['main_category'].tolist()

# Realizar la traducción en lotes utilizando el método translate_batch
translated_list = translator.translate_batch(text_list)

# Reemplazar la columna original por los valores traducidos
data['main_category'] = translated_list

# Imprimir el DataFrame con la columna traducida
print(data)
```
